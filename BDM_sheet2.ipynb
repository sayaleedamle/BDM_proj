```python
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i, y[i], y[i], ha = 'center')
```

```python
likes_2017 = df_likes.loc[df_likes['DATE'] < '2018-01-01']
print(likes_2017)
total_likes_2017 = likes_2017['LIKES'].sum()
print(total_likes_2017)
```

```python

plt.plot(likes_2017['DATE'], likes_2017['LIKES'])
```

```python
months=likes_2017['DATE'].dt.month
monthly_sum_2017=likes_2017.groupby(months).LIKES.sum()
monthly_sum_2017
l = list(monthly_sum_2017)
print(l)
m = set(months)
m = list(m)
m.sort()
print(m)
plt.xticks(np.arange(0,13))
plt.bar(m, l)
plt.xlabel('Month Number - 2017')
plt.ylabel('Number of likes')
```
